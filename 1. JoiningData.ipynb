```python
#Import packages
import numpy as np
import pandas as pd
```

```python
#Importing the raw .csv files
billamount = pd.read_csv('data/bill_amount.csv')
billid = pd.read_csv('data/bill_id.csv')
clinicaldata = pd.read_csv('data/clinical_data.csv')
demodata = pd.read_csv('data/demographics.csv')
```

```python
#Dsiplay basic information of the data
billamount.info()
print('\n')
billid.info()
print('\n')
clinicaldata.info()
print('\n')
demodata.info()
```

```python
#Check the no. of rows and columns in each data file
print('Shape of the Imported Datasets:')
print('Bill_Amount_Data', billamount.shape)
print('Bill_Id_Data', billid.shape)
print('Clinical_Data', clinicaldata.shape)
print('Demographics_Data', demodata.shape)
```

```python
#Merge 'Bill_Amount_Data' and 'Bill_Id_Data' using 'bil_id' column
data1 = pd.merge(billamount, billid, on = 'bill_id')
data1
```

```python
#Merge 'Clinical_Data' and 'Demographics_Data' using 'patient_id' column
data2 = pd.merge(clinicaldata, demodata, left_on = 'id', right_on = 'patient_id')
data2
```

```python
#Merge 'data1' and 'data2' using patient_id and date of admissions columns.
data3 = pd.merge(data1, data2, left_on = ['patient_id', 'date_of_admission'], right_on = ['id', 'date_of_admission'])
#Drop redundant column
data3 = data3.drop('id', axis=1).rename(columns = {'patient_id_x' : 'patient_id'})
data3
```

```python
#There are records with same patient id, date_of_admission and date_of_discharge but different bill_id
#We sum the bill id of such records by using groupby
grouped_data = data3.groupby(['patient_id', 'date_of_admission',
       'date_of_discharge'])['amount'].apply(lambda x : x.astype(int).sum()).reset_index()
```

```python
#drop the old amount column from data3 
data3.drop('amount', axis=1, inplace = True)
#drop records that has same patient_id, date_of_admission and date_of_discharge
data3.drop_duplicates(subset=['patient_id', 'date_of_admission','date_of_discharge'], keep='last', inplace=True)
joineddata = pd.merge(left=data3, right=grouped_data, right_index=True, on=['patient_id', 'date_of_admission','date_of_discharge'])
joineddata = joineddata.drop('bill_id', axis=1).reset_index().drop(columns=['index', 'patient_id_y'])
joineddata.to_csv('data/joined_data.csv', index = False)
joineddata
```
